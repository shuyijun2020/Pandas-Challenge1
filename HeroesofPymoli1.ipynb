```python


# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(purchase_csv)
purchase_df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Tot_purchases = purchase_df["Price"].count()
Sum_purchases = purchase_df["Price"].sum()
Uni_items = len(purchase_df["Item ID"].unique())
Ave_price = round(purchase_df["Price"].mean(),2)
purchase_analysis = pd.DataFrame({"Number of Unique Items":[Uni_items],"Average Price":[Ave_price],"Number of Purchases":[Tot_purchases],"Total Revenue":[Sum_purchases]})
purchase_analysis["Average Price"] = purchase_analysis["Average Price"].map('${:,.2f}'.format)
purchase_analysis["Total Revenue"] = purchase_analysis["Total Revenue"].map('${:,.2f}'.format)
purchase_analysis
```

* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 150]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39",">40"]
purchase_df["Age Groups"] = pd.cut(purchase_df["Age"],bins,labels=group_name)
purchase_df
```

```python
agegroup_purchasecount = purchase_df["Age Groups"].value_counts()
agegroup_purchasecount
```

```python
group_purchase = purchase_df.groupby(["Age Groups"])
agegroup_purchasetotal = group_purchase["Price"].sum()
agegroup_purchasetotal
```

```python
avgpurchase_price = agegroup_purchasetotal/agegroup_purchasecount
avgpurchase_price
```

```python
avgpurchase_person = agegroup_purchasetotal/age_total
avgpurchase_person
```

```python
agepurchasing_analysis = pd.DataFrame({"Purchase Count":agegroup_purchasecount, "Average Purchase Price":avgpurchase_price, "Total Purchase Value":agegroup_purchasetotal, "Avg Total Purchase per Person":avgpurchase_person})
agepurchasing_analysis['Average Purchase Price'] = agepurchasing_analysis['Average Purchase Price'].map('${:,.2f}'.format)
agepurchasing_analysis['Total Purchase Value'] = agepurchasing_analysis['Total Purchase Value'].map('${:,.2f}'.format)
agepurchasing_analysis['Avg Total Purchase per Person'] = agepurchasing_analysis['Avg Total Purchase per Person'].map('${:,.2f}'.format)
agepurchasing_analysis
```
